```python
# import matplotlib as mpl
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G = nx.Graph()
G.add_node(3, color="black")
G.add_node(5, color="blue")
G.add_node(7, color="yellow")
G.add_edge(3, 5, color="red", weight=2)
```

```python
# subax1 = plt.subplot(121)
# nx.draw(G, with_labels=True, font_weight='bold')
pos = nx.spring_layout(G)
edges = G.edges()
colors = [G[u][v]['color'] for u,v in edges]
n_colors = [G.nodes[n]['color'] for n in G.nodes()]
weights = [G[u][v]['weight'] for u,v in edges]
nx.draw(G, pos, with_labels=True, node_color= , edge_color=colors, width=weights)
# subax2 = plt.subplot(122)
# nx.draw_shell(G, nlist=[range(5, 10), range(5)], with_labels=True, font_weight='bold')
```
